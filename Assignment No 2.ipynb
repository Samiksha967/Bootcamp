```python
import pandas as pd

# Load the dataset
file_path = r"C:\Users\samik\Desktop\viit\bootcamp\Housing.csv"
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(df.head())
```

```python
# Histogram
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Distribution of Property Prices')
plt.xlabel('price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot

plt.figure(figsize=(10, 6))
sns.boxplot(x='bedrooms', y='price', data=df)
plt.title('Property Prices by Number of bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

```python
# Scatter Plot

plt.figure(figsize=(10, 6))
sns.scatterplot(x='area', y='price', data=df, hue='prefarea')
plt.title('Price vs Area')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
plt.show()
```

```python
# Bar Chart

plt.figure(figsize=(10, 6))
sns.barplot(x='prefarea', y='price', data=df, estimator='mean')
plt.title('Average Property Prices by Location Preference')
plt.xlabel('Location Preference')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
#  Pair Plot

sns.pairplot(df, vars=['price', 'area', 'bedrooms', 'bathrooms'], hue='prefarea')
plt.title('Pair Plot of Property Features')
plt.show()
```

```python
# Heatmap

# Strip whitespace from column names
df.columns = df.columns.str.strip()

# Step 1: Check the data types
print("Data Types:\n", df.dtypes)

# Step 2: Convert categorical variables to numeric (if needed)
# Example: Converting a 'Yes/No' column to 1/0
# Replace 'YourCategoricalColumn' with the actual column name
# df['YourCategoricalColumn'] = df['YourCategoricalColumn'].map({'yes': 1, 'no': 0})

# If you have more categorical columns, consider using pd.get_dummies
df = pd.get_dummies(df, drop_first=True)

# Step 3: Check the updated DataFrame
print("Updated DataFrame:\n", df.head())

# Step 4: Compute and plot the correlation heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
